```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
# Apply the fit_transform method on the label_encoder instance
customer_data_encoded = customer_data.apply(label_encoder.fit_transform)
print(customer_data_encoded)
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

# Assuming 'customer_data_encoded' is your DataFrame and 'Rate' is a numerical column
kmeans = KMeans(n_clusters=2)
kmeans.fit(customer_data_encoded) # Select the numerical column you want to use for clustering
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
x=customer_data_encoded.iloc[:,[2,4]].values
print(x)
```

```python
wcss=[]
for i in range(1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
sns.set()
plt.plot(range(1,11),wcss)
plt.title('The elbow point graph')
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=0)
y=kmeans.fit_predict(x)
print(y)
```

```python
plt.figure(1,figsize=(15,6))
n=0
for x in['Rate','Summary']:
  n+=1
  plt.subplots(1,n)
  plt.subplots_adjust(hspace=0.5,wspace=0.5)
  sns.distplot(customer_data_encoded[x],bins=15)
  plt.title('Distplot of {}'.format(x))
  plt.show()
```
